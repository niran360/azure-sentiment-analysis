```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv('Model_Analysis.csv')
df.head()
```

```python
df1 = df[['id', 'text_x', 'google_nlp', 'ibm_watson']]
```

```python
df1.head()
```

```python
df1['google_nlp_&'] = np.where(df1['google_nlp'] == df1['ibm_watson'], '', df1['google_nlp'])
df1.head()
```

```python
df1['&_ibm_watson'] = np.where(df1['google_nlp'] == df1['ibm_watson'], '', df1['ibm_watson'])
df1.head()
```

```python
df1['google_nlp_&'].replace('', np.nan, inplace=True)
```

```python
df1['&_ibm_watson'].replace('', np.nan, inplace=True)
```

```python
df1.dropna(subset=['google_nlp_&'], inplace=True)
```

```python
df1.dropna(subset=['&_ibm_watson'], inplace=True)
```

```python
df1.drop(['&_ibm_watson', 'google_nlp_&'], axis=1, inplace=True)
```

```python
df1.head()
```

```python
df1.to_csv('ibm-watson-drilldown.csv', index='id')
```

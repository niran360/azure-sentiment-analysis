>0.25 ‚Äî 1.0 positive
>-0.25 ‚Äî 0.25 neutral
>-1.0 ‚Äî -0.25 negative

```python
def sentiment (i):
    if (i >= 0.25) and (i <= 1.0):
        return 'positive'  
    elif (i>=-0.25) and (i<0.25):
        return 'neutral'
    else:
        return 'negative'
df['sentiment'] = df['sentiment/score'].apply(lambda x:sentiment(x))   
```

```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv('Model_Analysis.csv')
df = df.set_index('id')

df.head()
```

```python
df1 = df[['text_x', 'Azure_ta', 'ibm_watson']]
```

```python
df1.head()
```

```python
df1['Azure_ta_&'] = np.where(df1['Azure_ta'] == df1['ibm_watson'], '', df1['Azure_ta'])
df1.head()
```

```python
df1['&_ibm_watson'] = np.where(df1['Azure_ta'] == df1['ibm_watson'], '', df1['ibm_watson'])
df1.head()
```

```python
df1['Azure_ta_&'].replace('', np.nan, inplace=True)
```

```python
df1['&_ibm_watson'].replace('', np.nan, inplace=True)
```

```python
df1.dropna(subset=['Azure_ta_&'], inplace=True)
```

```python
df1.dropna(subset=['&_ibm_watson'], inplace=True)
```

```python
df1.drop(['&_ibm_watson', 'Azure_ta_&'], axis=1, inplace=True)
```

```python
df1.shape
```

```python
df1.to_csv('azur-ibm-watson-drilldown.csv', index='id')
```
